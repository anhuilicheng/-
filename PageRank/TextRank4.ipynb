# 1、展示textrank4zh模块的主要功能：
# 提取关键词
# 提取关键短语（关键词组）
# 提取摘要（关键句

```python
import codecs
# 从textrank4zh模块中导入提取关键词和生成摘要的类
from textrank4zh import TextRank4Keyword,TextRank4Sentence
```

```python
file=r'D:\NIO\NLP\CSDN.txt'
text=codecs.open(file,'r','GBK').read()
```

```python
#创建分词类的实例
tr4w=TextRank4Keyword()
#分析文本，窗口2，英文单词小写
tr4w.analyze(text=text,lower=True,window=2)
#输出
#前20个关键词
print('关键词：')
for item in tr4w.get_keywords(num=20,word_min_len=2):
    print(item.word,item.weight)#打印关键词和权重
print('\n')
#从关键短语列表中获取关键短语
print('关键短语为：')
for phrase in tr4w.get_keyphrases(keywords_num=20,min_occur_num=2):
    print(phrase)
print('\n')
```

```python
#创建分句类的实例
tr4s=TextRank4Sentence()
tr4s.analyze(text=text,lower=True,source='all filters')
print('摘要：')
#抽取3条句子作为摘要
for item in tr4s.get_key_sentences(num=3):
    #句子的索引，权重，内容
    print(item.index,item.weight,item.sentence)
```

# 展示textrank4zh模块的三种分词模式的效果
#    三种分词模式分别为：
#     words_no_filter模式：简单分词，不剔除停用词，不进行词性过滤
#     words_no_stop_words模式：剔除停用词
#     words_all_filters模式（默认）：即剔除停用词，又进行词性过滤

```python
tr4w=TextRank4Keyword()
tr4w.analyze(text=text,lower=True,window=2)
#将文本划分为句子
print('句子')
for s in tr4w.sentences:
    print(s)
print('\n')
```

```python
# 对句子列表中的句子进行分词，不进行词性过滤
print('word_no_filter:')
for words in tr4w.words_no_filter:
    print('/'.join(words))
print('\n')
```

```python
# 打印去掉停用词的词列表
print('word_no_stop_words:')
for words in tr4w.words_no_stop_words:
    print('/'.join(words))
print('\n')
```

```python
# 打印去掉停用词并进行词性过滤之后的词列表
print('word_all_filters:')
for words in tr4w.words_all_filters:
    print('/'.join(words))
print('\n')
```
